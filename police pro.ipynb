```python
#Q. 2) ( Based on Filtering + Value Counts ) - For Speeding , were Men or Women stopped more often ? 
df[(df['violation_raw']=='Speeding')|(df['violation']=='Speeding')].driver_gender.value_counts
```

```python
df[(df['violation']=='Speeding')].driver_gender.value_counts().plot(kind='bar')
```

```python
#Q. 3)Does gender affect who gets searched during a stop ?
df.groupby('driver_gender').search_conducted.sum().plot(kind='barh')
```

```python
#Q. 4) ( mapping + data-type casting ) - What is the mean stop_duration ?
df.stop_duration.value_counts().plot(kind='barh')
```

```python
df['stop_duration']=df['stop_duration'].map({'0-15 Min': 7.5,'16-30 Min': 23 , '30+ Min': 54})
df['stop_duration'].mean()
```
